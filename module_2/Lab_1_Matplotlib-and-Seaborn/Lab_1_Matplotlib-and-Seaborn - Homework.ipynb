#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here 
fitbit.head()
```

```python
for col in ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']:
    x = fitbit[col]
    y = fitbit['Steps']
    
    fig, ax = plt.subplots()
    
    plt.scatter(x,y)
    ax.set(title=col+" vs Steps", xlabel=col, ylabel="Steps")
        
    plt.show()
```

```python
x1 = fitbit['Minutes Lightly Active']
x2 = fitbit['Minutes Very Active']
x3 = fitbit['Minutes Sedentary']
y = fitbit['Steps']
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(15,5))
# sharey means "share y", aka whether all the y axis share the same values

ax1.set_title("Minutes Lightly Active vs Steps")
ax1.set_xlabel("Minutes Lightly Active")
ax1.set_ylabel("Steps")
ax1.plot(x1,y)

ax2.set(title="Minutes Very Active vs Steps", xlabel="Minutes Very Active", ylabel="Steps")
ax2.bar(x2,y)

ax3.set(title="Minutes Sedentary vs Steps", xlabel="Minutes Sedentary", ylabel="Steps")
ax3.scatter(x3,y)

plt.show()
```

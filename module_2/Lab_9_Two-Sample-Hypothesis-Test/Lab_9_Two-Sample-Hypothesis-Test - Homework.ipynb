# Challenge 2 - Matched Pairs Test

In this challenge we will compare dependent samples of data describing our Pokemon. Our goal is to see whether there is a significant difference between each Pokemon's defense and attack scores. Our hypothesis is that the defense and attack scores are equal. In the cell below, import the `ttest_rel` function from `scipy.stats` and compare the two columns to see if there is a statistically significant difference between them.

```python
# Your code here:

pokemon.head(10)
```

```python
mean_attack = pokemon['Attack'].mean()
mean_attack
```

```python
mean_defense = pokemon['Defense'].mean()
mean_defense
```

```python
std_attack = pokemon['Attack'].std()
std_attack
```

```python
std_defense = pokemon['Defense'].std()
std_defense
```

```python
samples_a = np.random.normal(mean_attack, std_attack, 1000)
samples_d = np.random.normal(mean_defense, std_defense, 1000)
```

```python
plt.hist(samples_a, alpha=.3, color='blue', bins=30, density=True)
plt.hist(samples_d, alpha=.3, color='red', bins=30, density=True)
```

```python
stats.ttest_rel(samples_a, samples_d).pvalue
```

If we observe a large p-value, for
example greater than 0.05 or 0.1 then we cannot reject the null
hypothesis of identical average scores. If the p-value is smaller
than the threshold, e.g. 1%, 5% or 10%, then we reject the null
hypothesis of equal averages.

Describe the results of the test in the cell below.

```python
# Your conclusions here:

# since the pvalue is much smaller than 0.05, we can reject the hypothesis that
# the defense and attack scores are equal.
```

# Bonus Challenge - The Chi-Square Test

The Chi-Square test is used to determine whether there is a statistically significant difference in frequencies. In other words, we are testing whether there is a relationship between categorical variables or rather when the variables are independent. This test is an alternative to Fisher's exact test and is used in scenarios where sample sizes are larger. However, with a large enough sample size, both tests produce similar results. Read more about the Chi Squared test [here](https://en.wikipedia.org/wiki/Chi-squared_test).

In the cell below, create a contingency table using `pd.crosstab` comparing whether a Pokemon is legenadary or not and whether the Type 1 of a Pokemon is water or not.

```python
# Your code here:

pokemon
```

```python
pokemon['Type bi'] = 'Not Water'
pokemon
```

```python
pokemon.loc[pokemon['Type 1']=='Water', 'Type bi'] = 'Water'
pokemon.head(15)
```

```python
samples_tp = np.random.choice(['Water', 'Not Water'], 90)
samples_lg = np.random.choice([False, True], 90)
```

```python
pd.DataFrame({'result1': samples_tp, 'result2': samples_lg})
```

Perform a chi-squared test using the `chi2_contingency` function in `scipy.stats`. You can read the documentation of the function [here](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.chi2_contingency.html).

```python
# Your code here:

crosstab = stats.contingency.crosstab(samples_tp, samples_lg)
count = crosstab[1]
```

```python
pvalue = stats.chi2_contingency(crosstab[1])[1]
pvalue
```

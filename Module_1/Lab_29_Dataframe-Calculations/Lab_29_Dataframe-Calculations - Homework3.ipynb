You are serving as a Data Analyst at the headquarter of an international fashion goods chain store. Your boss today asked you to do two things for her:

**First, identify two groups of customers from the data set.** The first group is **VIP Customers** whose **aggregated expenses** at your global chain stores are **above the 95th percentile** (aka. 0.95 quantile). The second group is **Preferred Customers** whose **aggregated expenses** are **between the 75th and 95th percentile**.

**Second, identify which country has the most of your VIP customers, and which country has the most of your VIP+Preferred Customers combined.**

## Q1: How to identify VIP & Preferred Customers?

We start by importing all the required libraries:

```python
# import required libraries
import numpy as np
import pandas as pd
```

Next, import `Orders` from Ironhack's database into a dataframe variable called `orders`. Print the head of `orders` to overview the data:

```python
# your code here
orders=pd.read_csv('Orders.csv')
orders
```

## How to label customers whose aggregated `amount_spent` is in a given quantile range?


We break down the main problem into several sub problems:

#### Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

#### Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?

#### Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?

*Note: If you want to break down the main problem in a different way, please feel free to revise the sub problems above.*

Now in the workspace below, tackle each of the sub problems using the iterative problem solving workflow. Insert cells as necessary to write your codes and explain your steps.

```python
# your code here
# Sub Problem 1: How to aggregate the  amount_spent for unique customers?
orders_agg = orders.groupby('CustomerID')[['amount_spent']].sum()
orders_agg
```

```python
orders_agg.rename(columns={'amount_spent': 'amount_spent_agg'}, inplace=True)
orders_agg
```

```python
# Sub Problem 2: How to select customers 
# whose aggregated amount_spent is in a given quantile range?
orders_agg['amount_spent'].quantile([0.75, 0.95])

#df[(df["Age"] >= 20) & (df["Age"] <= 30)]
```

```python
orders_agg.sort_values('amount_spent_agg', ascending=False)
```

```python
# Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?
```

```python
orders_agg.loc[orders_agg['amount_spent']>5840.182, 'customer_type'] = 'VIP Customer'
orders_agg.loc[(orders_agg['amount_spent']<=5840.182) & (orders_agg['amount_spent']>1661.640), 'customer_type'] = 'Preferred Customer'
orders_agg.loc[orders_agg['amount_spent']<=1661.640, 'customer_type'] = 'Normal Customer'
orders_agg
```

```python
orders_agg[orders_agg['customer_type']=='VIP Customer']
```

```python
orders_agg[orders_agg['customer_type']=='Preferred Customer']
```

Now we'll leave it to you to solve Q2 & Q3, which you can leverage from your solution for Q1:

## Q2: How to identify which country has the most VIP Customers?

```python
orders_merged = orders.merge(orders_agg, how="inner", on="CustomerID")
orders_merged
```

```python
orders_bycountry = orders_merged.groupby('Country')[['CustomerID']].nunique().sort_values('CustomerID', ascending=False)
orders_bycountry
```

```python
orders_bycountry['CustomerID'].sum()
```

```python
orders_merged
```

```python
orders_merged_vip = orders_merged[orders_merged['customer_type']=='VIP Customer']
orders_merged_vip
```

```python
orders_victory = orders_merged.groupby('Country')[['customer_type']].value_counts()
orders_victory.to_frame('total_number').reset_index()
```

## Q3: How to identify which country has the most VIP+Preferred Customers combined?

```python
# your code here
orders_victory.groupby('Country')[(orders_victory['customer_type']=='VIP Customer') | (orders_victory['customer_type']=='Preferred Customer')].counts()
```

## Extra exploration

```python
df['Amount'].quantile([0.25, 0.5, 0.75])
```

```python
df["Amount_categorical"]= "None"
df.loc[df["Amount"] <= 12500, "Amount_categorical"] = "Low"
df.loc[(df["Amount"] > 12500) & (df['Amount'] <= 19000), "Amount_categorical"] = "Medium Low"
df.loc[(df["Amount"] > 19000) & (df['Amount'] <= 30000), "Amount_categorical"] = "Medium High"
df.loc[df["Amount"] > 30000, "Amount_categorical"] = "High"
df
```

```python
pd.crosstab([df['Risk']], [df['Amount_categorical']], normalize=True)
pd.crosstab([df['Risk']], [df['Amount_categorical']], normalize=False)
```
